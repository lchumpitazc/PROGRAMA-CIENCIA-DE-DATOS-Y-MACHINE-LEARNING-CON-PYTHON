## **Ejercicio N° 01**

Fichero **autos.csv**

Generar una función y utilizar esta para recodificar la variable “país de
origen” (1, 2, 3) en el dataframe a través de la función apply.

```python
import pandas as pd

#importamos el fichero autos.csv
from google.colab import files
files.upload()

#Cargamos la data en el dataframe data_autos
data_autos = pd.read_csv('autos.csv', sep=";")

#Verificamos la data cargada en el dataframe
data_autos.head()
```

```python
#Nos piden recodificar la variable Pais_Origen (1,2,3)

#Verificamos primero si tenemos los 3 paises a codificar
data_autos["Pais_Origen"].value_counts()

#Los paises de origen son USA, Japan, Europe

#Asignaremos los codigos de la siguiente manera:
#1->USA
#2->Japan
#3-> Europe
```

```python
#Hacemos un backup de nuestro dataframe
df_autos = data_autos.copy()

#Utilizamos la funcion Apply() de nuestro dataframe
df_autos['Pais_Origen'] = df_autos['Pais_Origen'].apply(
#Utilizamos lambda para la recodificacion de la variable
    lambda x: 1 if x == 'USA' else(
              2 if x == 'Japan' else 3))

#Verificamos que los codigos se hayan asigando correctamente

print('variable Pais_Origen original\n', data_autos["Pais_Origen"].value_counts())
print(" ")
print("*"*50)
print(" ")
print('variable Pais_Origen recodificado\n', df_autos['Pais_Origen'].value_counts())
```

## **Ejercicio N° 02**

Generar un Notebook Colab con código y resultados, que contiene
información de autos de un periodo de tiempo. Resolver los siguientes:
- Limpieza y tratamiento de las variables (Según corresponda)
- Generar 5 nuevas variables aplicando los siguientes criterios con las
variables numéricas:
1. Log10(var1/var2)
2. Sqrt(var1)*exp(var2)/200
3. Si var 1 > var 2 → 5, caso contrario 3
4. 1/logn(var1/var2)*100
5. var2**2/var1

```python
#Tratamiendo de datos

#Cargamos nuestro dataframe 
df = data_autos

#Informacion del dataframe
print("Informacion de nuestro dataframe autos")
print(" ")
df.info()

#Verificamos la cantidad de datos nulos
print(" ")
print(" ")
print("*"*50)
print("Cantidad de nulos por columna")
print(" ")
df.isnull().sum().sort_values(ascending=False)

#Verificamos que tenemos 8 datos nulos en la variable Millas_Por_Galon y 6 datos nulos en la variable Caballo_De_Fuerza 
```

```python
#LIMPIEZA DE DATOS

#Llenamos nuestros datos vacios con el promedio

#variable Millas_Por_Galon
df['Millas_Por_Galon'] = df['Millas_Por_Galon'].fillna(df['Millas_Por_Galon'].mean())

#variable Caballo_De_Fuerza
df['Caballo_De_Fuerza'] = df['Caballo_De_Fuerza'].fillna(df['Caballo_De_Fuerza'].mean())

#Verificamos la cantidad de datos nulos
print(" ")
print("Cantidad de nulos por columna")
print(" ")
df.isnull().sum()

#Verificamos nuestra data de cada variable sin nulos
```

```python
df.head()
```

```python
#GENERACION 5 NUEVAS VARIABLES
import numpy as np

#Definimos nuestras variables var1 y var2
var1 = df['Millas_Por_Galon']
var2 = df['Caballo_De_Fuerza']


#Variable Log10(var1/var2)
div = var1/var2
log10 = np.log10(div)
df['varLog10']= log10

#Variable Sqrt(var1)*exp(var2)/200
var_sqrt = (np.sqrt(var1)**var2)/200
df['varSqrt'] = var_sqrt

#Variable var 1 > var 2 → 5, caso contrario 3
var_cond = np.where(var1>var2,5,3)
df['varCond'] = var_cond

#Variable 1/logn(var1/var2)*100
varLogn = (1/np.log(div))*100
df['varLogn'] = varLogn

#Variable var2**2/var1
varExpo = (var2**2)/var1
df['varExpo'] = varExpo

#Verificamos las 5 variables en nuestro dataframe
df.iloc[:, 9:14].head()
```

## **Ejercicio N° 03**

Crear una función que permita generar 3 variables cualitativas
ordinales a partir de las variables numéricas con los siguientes criterios:

★ Si es mayor a la “media + 1*desviación estándar” → Alto

★ Si está entre “media - 1*desviación estándar” y “media +1*desviación estándar” → Medio

★ Si es menor a la “media - 1*desviación estándar” → Bajo

Nota: Mostrar con las nuevas variables generadas.
Elegir con libertad las variables a trabajar.
La función debe contar con su respectiva documentación.

```python
import statistics as stats

#creamos la funcion var_cualitativa
def var_cualitativa(a:str, b:pd.DataFrame)->str:
  
  #Establecemos nuestros rangos superior e inferior
  limite_superior = df[a].mean() + 1*df[a].std()
  limite_inferior = df[a].mean() - 1*df[a].std()
  
  #Con lambda definimos los rango (Alto,Medio,Bajo)
  return lambda x: "Alto" if x>limite_superior else("Medio" if(x>=limite_inferior and x<=limite_superior) else "Bajo")

#Utilizamos la funcion con la variable Millas_Por_Galon y definimos nuestra nueva columna Rango en el dataframe
df['Rango'] = df['Millas_Por_Galon'].apply(var_cualitativa(a = 'Millas_Por_Galon', b=df))

#Verificamos la columna Rango
df.iloc[:, 9:15].head()
  
```

## **Ejercicio N° 04**

Crear una función que muestre los siguientes gráficos
(simultáneamente) para las variables numéricas, según una variable
cualitativa:

❖ Cajas y bigotes

❖ Barras de la media

❖ Histograma con densidad (Curva)

```python
#importamos nuestras librerias
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import missingno as msno
```

```python
#FUNCION CAJA Y BIGOTE
def graphic_boxplot (var_df,var_x,var_y):
  #en el dataframe filtramos nuestra variables
  var_df=var_df.loc[:,[var_x,var_y]]
  #definimos el tamaño del grafico
  fig, ax = plt.subplots(figsize=(8, 4))
  #usamos la funcion boxplot para el grafico
  sns.boxplot(
        x     = var_x,
        y     = var_y, 
        data  = df,
        ax    = ax
    )
  #definimos un titulo para el grafico
  ax.set_title('Diagrama de caja y bigotes');
  return

#FUNCION GRAFICO BARRAS CON LA MEDIA
def graphic_bar(var_df,var_x,var_y,var_values):
  #usamos la funcion crosstab para nuestras variables x y, definimos la media, y el valor de la variable para la media
  df_graphic_bar = pd.crosstab(var_df[var_x],var_df[var_y],aggfunc='mean',values=df[var_values]).round(2)
  #usamos la funcion del grafico
  df_graphic_bar.plot.bar()
  #mostramos el grafico
  plt.show()
  return

#FUNCION HISTOGRAMA CON DENSIDAD DE CURVA
def graphic_hist(var_df,var):
  #usamos la funcion del grafico histplot, definimos el dataframe, la variable, y la densidad de la curva (kde)
  sns.histplot(data = df,x = df[var],color = 'navy',kde = True)
  #mostramos el grafico
  plt.show()
  return

#usamos nuestras funciones para mostrar los graficos
graphic_boxplot(df,'Rango','Millas_Por_Galon')
graphic_bar(df,'Pais_Origen','Rango','Millas_Por_Galon')
graphic_hist(df,'Rango')
```

## **Ejercicio N° 05**

Crear una función que devuelva un mapa de calor con la correlación
de todas las variables numéricas (incluye las generadas), para cada
país.

```python
#Creamos nuestra funcion de mapa de calor
def Mapa_Calor(var_df):
  #seleccionamos la variable Pais_Origen
  df_paises = var_df['Pais_Origen']
  #quitamos los duplicados
  df_paises.drop_duplicates()
  #lo pasamos a un array con numpy
  array_paises = df_paises.to_numpy()
  #lo pasamos a una lista
  listado_paises = array_paises.tolist()
  #lo seteamos para luego iretarlos
  set_paises = set(listado_paises)

  #usamos un for para mostrar cada mapa de calor
  for i in set_paises:
    i_paises = var_df['Pais_Origen'] == i
    df_paises = df[i_paises]
    #definimos el tamaño del grafico
    plt.figure(figsize=(8,6))
    #definimos el titulo del grafico
    plt.title("Mapa de calor de "+i)
    #usamos la funcion del dataframe correlacion
    correlacion = df_paises.corr()
    #usamos la funcion del grafico heatmap(mapa de calor)
    sns.heatmap(correlacion,
            xticklabels = correlacion.columns.values,
            yticklabels = correlacion.columns.values,
            annot = True,
            cmap = "hot"
            );
  return
#usamos nuestra funcion para mostra el grafico
Mapa_Calor(df)
```
